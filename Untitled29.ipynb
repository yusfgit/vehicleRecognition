```python
# Install Ultralytics for YOLO Train
!pip install ultralytics

# import ultralytics
from ultralytics import YOLO

#Select Yolo Model
model = YOLO('yolov8n.pt')

#.yaml file path (It has train path/validation path/classNumber/classes infos) 
data_yaml_path = '/content/vehicle_dataset/dataset/vehicle_dataset/data.yaml'

# Train Model
model.train(data=data_yaml_path, epochs=50)
```
